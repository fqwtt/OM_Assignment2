```python
from pyomo.environ import *
import pandas as pd 
import numpy as np
from math import log,e
from sklearn.linear_model import LinearRegression

```

```python
Beer_data_12pk = pd.read_excel('BeerSales_Data.xlsx',sheet_name='12PK')
Beer_data_18pk = pd.read_excel('BeerSales_Data.xlsx',sheet_name='18PK')
Beer_data_30pk = pd.read_excel('BeerSales_Data.xlsx',sheet_name='30PK')
Beer_data_a= pd.read_excel('BeerSales_Data.xlsx',sheet_name='a')
```

```python
Price = { '12_packs_size': [20.99, 18.99, 16.99, 14.99],
          '18_packs_size': [19.99, 17.99, 15.99, 13.99],
          '30_packs_size': [18.99, 16.99, 15.49, 13.49]  
}
```

```python
Week = range(53,66)
packs_size=['12_packs_size','18_packs_size','30_packs_size']
n=64
m=16
df=np.zeros((n,m))
WR_table=pd.DataFrame(df,columns=['12','18','30']+['Week'+str(i) for i in Week])
```

```python
Beer_data_a
```

```python
#ln
a=np.array(Beer_data_a.iloc[:,1:])
for week in Week:
    for i in range(len(Price[packs_size[0]])):
        P12=Price[packs_size[0]][i]
        for j in range(len(Price[packs_size[1]])):
            P18=Price[packs_size[1]][j]
            for k in range(len(Price[packs_size[2]])):
                P30=Price[packs_size[2]][k]
                row=i*16+j*4+k
                WR_table.iloc[row,0]=P12
                WR_table.iloc[row,1]=P18
                WR_table.iloc[row,2]=P30
                WR=0
                WR+=P12*pow(e,a[0][0]+a[0][1]*week+a[0][2]*log(P12)+a[0][3]*log(P18)+a[0][4]*log(P30))
                WR+=P18*pow(e,a[1][0]+a[1][1]*week+a[1][2]*log(P12)+a[1][3]*log(P18)+a[1][4]*log(P30))
                WR+=P30*pow(e,a[2][0]+a[2][1]*week+a[2][2]*log(P12)+a[2][3]*log(P18)+a[2][4]*log(P30))
                WR_table.iloc[row,week-50]=WR
```

```python
Week = range(53,66)
Price = range(64)
```

```python
model = ConcreteModel()
model.x = Var(Week, Price , within=Binary)

model.obj = Objective(
    expr = sum(model.x[week,price]*WR_table.iloc[price,week-50] for week in Week for price in Price),
    sense = maximize)

def rule1(model,week):
    return sum(model.x[week,price] for price in Price)==1

model.con1 = Constraint(Week,rule=rule1)

opt = SolverFactory('glpk').solve(model)

print('The optimal objective value is: %f'%model.obj())
```

```python
result=pd.DataFrame(index=['Week'+str(i) for i in Week],columns=['P12','P18','P30','WR'])
for week in Week:
    for price in Price:
        if model.x[week,price].value==1:
            p12=WR_table.iloc[price,0]
            p18=WR_table.iloc[price,1]
            p30=WR_table.iloc[price,2]
            # print('In week %d, the optimal price is: 12 packs: %.2f, 18 packs: %.2f,30 packs: %.2f'%(week,p12,p18,p30))
            result.iloc[week-53,0]=p12
            result.iloc[week-53,1]=p18
            result.iloc[week-53,2]=p30
            result.iloc[week-53,3]=WR_table.iloc[price,week-50]
```
